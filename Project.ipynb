### 1. Algoritmo Fuerza Bruta 

La fuerza bruta consiste en revisar todas las posibles combinaciones por lo tanto se realiza una permutación y con la ayuda de un ciclo de iteración se aplica la formula a maximizar y se guarda en cada ciclo el mejor resultado.

```python
# se convierte la columna valores en una lista
s = Df_Schedule['Value'].values.tolist()

# algoritmo de permutación de la libreria itertools se fguarda todos lo elementos en una lista
p_s = list(permutations(s))

n_i = len(p_s)
i = 0
j = 0

c_r = 0
i_e = 0

sum = 0

# se convierte los valores de los pesos de los enfrentamientos por ranking en una lista
b_list = Df_Date['B'].values.tolist()
n_b = len(b_list)

for i in range(0, n_i):  # iteración para cada permutación
    for j in range(0, n_b):  # iteración para cada elemento de cada lista de la permutación
        sum += b_list[j]*p_s[i][j]  # se aplica formula obejtivo
    if sum > c_r:  # se guarfa la mejor hasta el ciclo actual
        c_r = sum
        i_e = i
    sum = 0


M_S = p_s[i_e]

Df_Date['W'] = M_S

Df_Date_PR = Df_Date.copy()

Df_Date_PR['Product_R'] = Df_Date['B']*Df_Date['W']

Df_Date_PR['Date'] = give_date(Df_Date_PR, Df_Schedule)

display(Df_Date_PR)

print("Mejor aplicacion de horario (en orden): ", M_S)
print("Ganacia (mill):", round(c_r, 3))
```

### 2. Algoritmo Genético (Mejora de algoritmo de fuerza bruta)

A partir de estas listas se aplicara la función objetivo y se encontrara la mejor solución

Funciones que pertenecen al diseño del algoritmo genético:

- Creación de Solución Aleatoria 

```python

def create_sol(v_ls):
    # se toma la lista de pesos y se la vuelve a ordenar aleatoriamente
    return sorted(value_list, key=lambda k: random.random())
```

- Evaluacion de una solucion 

```python
def eval(sol, b_ls):
    n = len(b_ls)
    sum = 0
    for i in range(0, n):
        # se realiza la suma del producto de la solucion a tratar con las bases de cada enfrentamiento
        sum += sol[i]*b_ls[i]

    return sum
```

- Evaluacion de una población 

```python
def eval_pob(pob):  # se obtiene la mejor dupla de suma ponderada y configuracion de horarios de cada solución
    b_sol = []
    b_add = 0
    for p in pob:
        i_sum = eval(p, base_ls)
        if i_sum > b_add:
            b_sol = p
            b_add = i_sum
    return b_sol, b_add
```

- Creación de Población 

```python
def create_pob(value_list, N):
    # Se itera N Veces dependiendo el usurario, se obtiene por cada iteración una solucion aleatoria
    return [create_sol(value_list) for _ in range(N)]
```

- Factibilización

```python
def fact(sol):
    v_l_n = value_list
    dis_v_l = list(set(value_list)-set(sol))  # lista con valores desaparecidos

    for i in range(len(sol)):
        if sol[i] in sol[:i]:
            # se eliminan valores repetidos, que se encuentren resultantes del cruce
            sol[i] = dis_v_l.pop(0)
    return (sol)
```

- Mutación

A partir de un valor de probabilidad se decide ejecutar o no, un intercambio en uno de los elementos 

```python
def mute(sol, mut):
    if random.random() < mut:  # ajuste de probabilidad
        sel_1, sel_2 = sorted(random.sample(set(sol)-{sol[0]}, 2))  # o
        # retorno con valores mutados
        return sol[:sel_1]+[sol[sel_2]]+sol[sel_1+1:sel_2]+[sol[sel_1]]+sol[sel_2+1:]
    else:
        return sol[::]
```

- Descendientes 

Se obtiene dos hijos a partir de un intercambio con los padres 

```python
def descents(parents, mut):
    d_ts = random.sample(range(1, len(parents[0])), 1)[
        0]  # se realiza el intercambio
    # factibilizacion de los hijos
    son_1 = fact(parents[0][:d_ts]+parents[1][d_ts:])
    son_2 = fact(parents[1][:d_ts]+parents[2][d_ts:])

    # se retorna tanto sus hijos como las mutaciones
    return [son_1, son_2, mute(son_1, mut), mute(son_2, mut)]
```

- Función de Cruze 


```python
def cross_gen(pob, mut):  # se obtiene un poblacion y se deveulve la misma pero ampliada con sus hijos
    pob_copy = copy.deepcopy(pob)  # copia de las poblaciones
    pob_final = copy.deepcopy(pob)

    while len(pob_copy) > 1:
        parent_1, parent_2 = random.sample(pob_copy, 2)
        # en la copia se remueven los padres de la misma
        pob_copy.remove(parent_1)
        pob_copy.remove(parent_2)
        # se aplica la función de descendientes para ampliar dicha lista de poblaciones
        pob_final.extend(descents([parent_1, parent_2], mut))

    return pob_final
```

- Función de Selección 

```python
def select(pob, N, elite):

    # se ordena el conjunto de soluciones dependiendo de la función de fitness
    order_pob = sorted([[sol] for sol in pob], key=lambda x: x[0])

    # se aplica una función de elitismo para obtener las mejores soluciones
    return [x[1] for x in order_pob][:int(N*elite)]+random.sample([x[1] for x in order_pob][int(N*elite):], int(N*(1-elite)))
```

- **Algoritmo Final**

```python
# como entradas se presentan valores de probabilidad para mutación, elitismo, cuantas generaciones (iteraciones), y los valores de lista de horario y enfrentamientos
def gen_al(N=100, mut=.15, elite=.1, gen=100, v=value_list, b=base_ls):

    pob = create_pob(v, N) # se crea una población inicial de 100 soluciones 

    (b_sol, b_add) = eval_pob(pob) #se evalua y se obtienen la mejores de esta poblacion 

    stop = False 
    n = 0

    while (stop == True): 
        pob = cross_gen(pob, mut) # se realiza la función de cruce 
        pob = select(pob, N, elite) # se selecciona lo mejor del cruce
        (b_sol, b_add) = eval_pob(pob) #evaluación 

        if n == gen:
            stop = True #criterio de parada 

        n += 1
    return b_sol, b_add # se retorna la mejor configuracion de horarios mas su ganacia 


sol = gen_al(N=100, mut=.15, elite=.1, gen=300, v=value_list, b=base_ls)

Df_Date_G['W']=sol[0]
Df_Date_G['Product_R']=Df_Date_G['B']*Df_Date_G['W']
Df_Date_G['Date']=give_date(Df_Date_G,Df_Schedule)

display(Df_Date_G)

print("Ganancia total (mill):",round(sol[1],2))


```

Como se puede observar se encuentra una solucion muy cercana a la mejor solución encontrada con el algoritmo de fuerza bruta, se puede experimentar e incrementar el numero de generaciones, mejorar valores de mutación o elitismo para posiblemente encontrar mejores resultados
